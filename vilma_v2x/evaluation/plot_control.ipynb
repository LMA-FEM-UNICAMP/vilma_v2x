```python
import pandas as pd

import matplotlib.pyplot as plt

import numpy as np

import scienceplots

plt.style.use(['science','ieee'])

df = pd.read_csv('out0.csv')

df2 = pd.read_csv('out1.csv')

df3 = pd.read_csv('out2.csv')

print(df.columns)
print(df2.columns)
print(df3.columns)

```

```python
t = df['t'].tolist()
t2 = df2['t'].tolist()
t3 = df3['t'].tolist()
out0 = df['out0'].tolist()
out1 = df2['out1'].tolist()
out2 = df3['out2'].tolist()

ref = [35.0 for i in range(len(t))]
```

```python
plt.figure(figsize=(4, 3))

plt.plot(t, ref, label=r'Setpoint')
plt.plot(t, out0, label=r'$\sigma = 0.00$')
plt.plot(t2, out1, label=r'$\sigma = 0.42$')
plt.plot(t3, out2, label=r'$\sigma = 3.15$')
plt.xlabel('Time (s)')
plt.ylabel('Distance (m)')
plt.xlim(t[0],t[len(t)-1])
plt.title(f'Platooning simulation')
plt.legend(loc='lower right')
plt.savefig('platooning_control.pdf', format='pdf')
plt.show()
```
